```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('/content/transaction_dataset.csv')

gender_count = dataset['Gender'].value_counts(dropna=False)
gender_count.plot(kind='bar', color='skyblue')
plt.title('Customer gender count')
plt.xlabel('Gender')
plt.ylabel('Count')
```
